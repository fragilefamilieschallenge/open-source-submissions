```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn_pandas import DataFrameMapper

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
features = pd.read_csv('features_imputed_set1.csv', low_memory=False, index_col='challengeID')
print("Features shape: {}".format(features.shape))
features.sort_index(kind='mergesort', inplace=True) # mergesort is the only stable sort
features.head()
```

```python
# scale data to unit variance and 0 mean
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
indices_f = features.index
colnames_f = features.columns
features = ss.fit_transform(features)
features = pd.DataFrame(data=features, index=indices_f, columns=colnames_f)
features.head()
```

```python
from sklearn.model_selection import train_test_split

targets = pd.read_csv('train.csv', index_col='challengeID')
full_y = targets['gpa'] # we are specifically interested in GPA
print("Target shape: {}".format(full_y.shape))
full_y.dropna(how='any', inplace=True) # drop those with no reported GPA
print("Target shape (no na): {}".format(full_y.shape))

# get the rows for which we can predict
y_indices = full_y.index.values.tolist()
full_x = features.loc[y_indices]
print("Features shape (final): {}".format(full_x.shape))

train_x, test_x, train_y, test_y = train_test_split(
    full_x, full_y, test_size=0.2, random_state=0)

train_x.shape
train_x.head()
train_y.shape
train_y.head()
test_x.shape
test_x.head()
test_y.shape
test_y.head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.kernel_ridge import KernelRidge
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_squared_error

# Test on
# Lasso - Linear Regression
# Lasso
# Random Forest/ Decision Tree
# SVR
# Kernel Ridge Regression?
# 

#clf - the classifier, params - params to CV, transform - to do any feature selection
classifiers = [
    {
        'clf': DecisionTreeRegressor(),
        'params': {
            "max_depth": [20, None],
            "max_features": [None],
            "criterion": ["mse", "mae"]
        }
    },
    {
        'clf': RandomForestRegressor(),
        'params': {"max_depth": [20, None],
                  "max_features": [None],
                  "criterion": ["mse", "mae"]}
    },
    {
        'clf': Lasso(),
        'params': {
            "alpha": [1, 2, 4, 8],
            "fit_intercept": [True, False],
            "normalize": [True, False]
        }
    },
    {
        'clf': LinearRegression(),
        'params': {
            "fit_intercept": [True, False],
            "normalize": [True, False]
        }
        
    },
#     {
#         'clf': KernelRidge(),
#         'params': {
            
#         }
        
#     },
    {
        'clf': LinearSVR(),
        'params': {
            "C": [1, 2, 4]
        }
        
    }
]

# run grid search without feature selection
for clf in classifiers: 
    model = clf['clf']
    print("~~~ Fitting and Testing {} ~~~".format(type(model).__name__))
    grid_search = GridSearchCV(model, param_grid=clf['params'])

    grid_search.fit(train_x, train_y)
    grid_search.cv_results_

    clf = grid_search.best_estimator_
    clf.fit(train_x, train_y)
    pred_y = clf.predict(test_x)

    "----MSE: {}----".format(mean_squared_error(test_y, pred_y))
```
